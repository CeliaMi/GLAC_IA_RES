```python
# Uniremos los dos datasets en base a que los dos primeros d√≠gitos de la columna latitud y longitud coincidan junto al a√±o
# Para ello a√±adiremos columnas en ambos datsets con solo los dos primeros d√≠gitos.

# Funci√≥n para obtener los primeros dos caracteres si es positivo, tres si es negativo
def obtener_primeros_digitos(valor):
    valor_str = str(valor)
    if valor_str.startswith('-'):
        return valor_str[:2]
    else:
        return valor_str[:1]

temp_df['longitud_2d'] = temp_df['Longitude'].apply(obtener_primeros_digitos)
temp_df['latitud_2d'] = temp_df['Latitude'].apply(obtener_primeros_digitos)
copy_prueba['longitud_2d'] = copy_prueba['LONGITUDE'].apply(obtener_primeros_digitos)
copy_prueba['latitud_2d'] = copy_prueba['LATITUDE'].apply(obtener_primeros_digitos)
```
